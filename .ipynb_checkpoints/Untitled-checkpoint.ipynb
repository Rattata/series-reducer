```julia
table = readcsv("D:\\lab\\series-reducer\\6df85afc-36cb-46c9-98dd-dd6f70322d551117945382446426219.csv"; skipstart=1)
```

```julia
values = Array{Float64}((600*12),3)
for r in 1:600
    for c in 2:13
        a = [table[r,1],Float64(c - 1),table[r,c]]
        values[(r - 1) * 12 + (c - 1), :]  = a
    end
end
values = DataFrame(values)
```

### Linear regression
As you can see below, increases in N confirm the algorithm has a time complexity of O(n)

```julia
model = glm(@formula(x3 ~ x2 + x1), values, Normal(), IdentityLink())
```

```julia
valuesmedian = Array{Float64,2}(0,3);
for i in unique(values[:,:N])
    for j in unique(values[:,:threads])
        subset = values[(values[:N].==i).&(values[:threads].==j),:]
        valuesmedian = vcat(valuesmedian,[i , j, mean(subset[:time_ns])]')
    end
end
valuesmedian
```
